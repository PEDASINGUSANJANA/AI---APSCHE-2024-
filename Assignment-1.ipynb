##BASIC PYTHON

Task 1: Write a Python program to check if a number is even or odd.
The program should take an integer input from the user and print whether the number is even or odd.

```python
n=int(input("Enter a number:"))
if n%2==0:
  print(n,"is even")
else:
  print(n,"is odd")
```

Task 2: Write a Python program to find the largest of three numbers.
The program should take three integer inputs from the user and print the largest number.

```python
A=int(input("Enter A:"))
B=int(input("enter B:"))
C=int(input("enter C:"))
if A>B and A>C:
  print("The largest number is",A)
elif B>A and B>C:
    print("The largest number is",B)
else:
    print("The largest number is ",C)
```

##Matplotlib

Task 1: Create a line plot using Matplotlib.
Plot the function y = x^2 for x values ranging from -10 to 10. Label the axes and give the plot a title.

```python
import matplotlib.pyplot as plt
x=np.linspace(-10,10,5)
y=x**2
```

```python
plt.plot(x,y)
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('lineplot of y=x^2')
plt.show()
```

Task 2: Create a bar chart using Matplotlib.

```python
import matplotlib.pyplot as plt
subjects = ['Mathematics', 'DS', 'English', 'AI', 'DBMS']
average_marks = [85, 78, 90, 72, 88]
```

```python
plt.figure(figsize=(10, 6))
plt.bar(subjects, average_marks, color='blue')
plt.xlabel('Subjects')
plt.ylabel('Average Marks')
plt.title('Average Marks in Different Subjects')
for i, v in enumerate(average_marks):
    plt.text(i, v + 1, str(v), ha='center', va='bottom')
plt.show()
```

Task 2: Create a heatmap using Seaborn.

```python
import seaborn as sns
data = {
    'age': [19, 18, 28, 33, 32],
    'bmi': [27.9, 33.77, 33.0, 22.705, 28.88],
    'children': [0, 1, 3, 0, 0],
    'charges': [16884.924, 1725.5523, 4449.462, 21984.47061, 3866.8552]
}
df = pd.DataFrame(data)
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True, square=True, fmt='.2f')
plt.title('Correlation Matrix Heatmap')
plt.show()
```
